```
import sklearn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn_pandas import DataFrameMapper, cross_val_score
import sklearn.preprocessing, sklearn.decomposition, sklearn.linear_model, sklearn.pipeline, sklearn.metrics
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```
df = pd.read_csv('./Earthquake-Forecasting/final_data.csv')
```

```
mapper = DataFrameMapper([
        (['place'], None),
        (['mag'],sklearn.preprocessing.StandardScaler()),
        (['latitude'],sklearn.preprocessing.StandardScaler()),
        (['longitude'],sklearn.preprocessing.StandardScaler()),
        (['depth'],sklearn.preprocessing.StandardScaler()),
        (['timestamp'], sklearn.preprocessing.StandardScaler()) 
    ],default = None)
```

```
pipe = sklearn.pipeline.make_pipeline(mapper, sklearn.svm.SVC())
```

Making the problem a supervised learning problem

```
X = df.copy()
del X['mag']
del X['place']
y = df.mag
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```
svm = sklearn.svm.SVR()
svm.fit(X_train, y_train)
```

Obtaining the **MSE** (Mean Squared Error)

```
from sklearn.metrics import mean_squared_error
mean_squared_error(svm.predict(X_test), y_test)
```

```
for a,b in zip(svm.predict(X_test), y_test):
    print(a,b)
```
